```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.load('Er170_YSO_Freespace_-15dBm.v2.npz')
```

```python
df = pd.DataFrame()
df = df.assign(laser_wavelength = data['laser_lambdas'])
df = df.assign(wavelength = data['lambdas'])
df = df.assign(power = data['power'])
```

```python
min(data['power'])
```

```python
df = df.pivot(index = 'laser_wavelength', columns = 'wavelength', values = 'power')
```

```python
ax = sns.heatmap(df,vmax = -80)
ax.invert_yaxis()
```
