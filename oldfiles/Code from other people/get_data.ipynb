```python
import numpy as np
from numpy import load
import matplotlib
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import Axes3D
import math
from matplotlib.collections import PolyCollection
from matplotlib import colors as mcolors
```

```python
data = np.load("Freespace-20240923_new_GRIN_Groove_with_YSO_crystal_-10_dBm.npz")
```

```python
pm_range = int(data['PM_RANGE'])
watt = list(map(lambda x: x*10**(pm_range/10)*1e-3,data['data_arr_cut']))
plt.plot(data['wav_arr'],watt)
plt.title('Freespace coupling through YSO crystal room temperature')
plt.xlabel('wavelength (nm)')
plt.ylabel('power(W)')
plt.show()
```

```python
def smallest(lst):
    track = 10000
    for i in lst:
        if i>0 and i<track:
            track = i
    return track
s = smallest(watt)
dBm = list(map(lambda x: 10*math.log10(x)+30,watt))
plt.plot(data['wav_arr'],dBm)
plt.title('Freespace coupling through YSO crystal room temperature 0dBm')
plt.xlabel('wavelength(nm)')
plt.ylabel('power(dBm)')
plt.show()
```
